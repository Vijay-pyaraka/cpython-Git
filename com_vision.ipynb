**Scalling**

```python
trans_scal = np.array([[2,0,0],[0,2,0]])
T=trans_scal.astype(float)
```

```python
L= cv2.warpAffine(flower_resize,T,(200,200))
```

```python
cv2_imshow(L)
```

**Translation**

```python
translate_mat=np.array([[1,0,50],[0,1,100]]).astype(float)
```

```python
Traslation = cv2.warpAffine(flower_resize,translate_mat,(200,200))
```

```python
cv2_imshow(Traslation)
```
